```python
import pandas as pd 
import numpy as np
from numpy.random import randn
%matplotlib inline 

# Create DataFrame
my_df = pd.DataFrame(randn(500, 4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
# Density and Kernal Density Estimation plots with scipy library 
my_df.plot(kind="kde")
```

```python
# KDE second way to learn shift tab 
my_df.plot.kde()
```

```python
# Density plot 
my_df.plot.density()
```

```python
# Other way 
my_df.plot(kind="density")
```

```python
# Change stuff
my_df["Mon"].plot(kind="density", alpha=0.6, cmap="magma", lw=4)
```

```python
# title & grid
my_df.plot(kind="kde", title="My Awesome KDE plot!", grid=True)
```
