```python
import pandas as pd 
import numpy as np
from numpy.random import randn
%matplotlib inline
```

```python
# Create DataFrame 
my_df = pd.DataFrame(randn(25, 4), columns = ["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
# bar plot
my_df.plot(kind = "bar")
```

```python
# bar plot(absolute)
my_df.abs().plot(kind = "bar")
```

```python
# bar plot stacked 
my_df.abs().plot(kind = "bar", stacked = True)
```

```python
# bar plot stacked & grid 
my_df.abs().plot(kind = "bar", stacked = True, grid = True)
```

```python
# Shaddig 
my_df.abs().plot(kind = "bar", stacked = True, alpha=0.3)
```

```python
# Title 
my_df.abs().plot(kind = "bar", stacked = True, title="My Awesome Bar Chart")
```

```python
# Legend removal 
my_df.abs().plot(kind = "bar", stacked = True, title="My Awesome Bar Chart", legend=False)
```

```python
# Second Way
my_df.abs().plot.bar()
```

```python
# Second Way Stacked 
my_df.abs().plot.bar(stacked=True)
```

```python
# Second Way Shadding 
my_df.abs().plot.bar(stacked=True, alpha=0.4)
```

```python
# Second Way Title 
my_df.abs().plot.bar(stacked=True, title="My Bodacious Bar Chart")
```

```python
# Second Way Legend 
my_df.abs().plot.bar(stacked=True, title="My Bodacious Bar Chart", legend=False)
```

```python
# # Second Way Grid Line 
my_df.abs().plot.bar(stacked=True, title="My Bodacious Bar Chart", legend=True, grid=True)
```
