```python
import pandas as pd 
import numpy as np
from numpy.random import randn
%matplotlib inline 

# Create DataFrame
my_df = pd.DataFrame(randn(500, 4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
# Boxplot
my_df.plot(kind="box")
```

```python
# Boxplots Size
my_df.plot(kind="box", figsize=(10, 7))
```

```python
# Boxplot color : https://matplotlib.org/stable/users/explain/colors/colormaps.html
my_df.plot(kind="box", cmap="hsv")
```

```python
# No Legend
my_df.plot(kind="box", legend=True)
```

```python
# Title 
my_df.plot(kind="box", title="My Awesome Box Plot!")
```

```python
# Other Way
my_df.plot.box()
```

```python
# Adding Grid 
my_df.plot.box(title="My Awesome Box Chart!", grid=True)
```
