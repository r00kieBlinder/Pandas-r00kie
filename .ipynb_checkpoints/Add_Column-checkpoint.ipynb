```python
import pandas as pd 
import numpy as np
```

```python
# import csv file
my_df = pd.read_csv('dog_data_short.csv')
my_df
```

```python
# Add column from List
gender = ["Male", "Female", "Female", "Male", "Male"]
my_df["Gender"] = gender
my_df
```

```python
# Add Default Values 
my_df["Alive/Dead"] = [True]*len(my_df)
my_df
```

```python
# Add NaN value -> NaN(Not a Number) -> null value 
my_df["Show Dog"] = [np.nan]*len(my_df)
my_df
```

```python
# add columns with .insert() - Allows Position
my_df.insert(3, "Adopted", [True]*len(my_df), True)
my_df
```

```python
# add column with .assign() - creates new df
my_df2 = my_df.assign(Horse=[False]*len(my_df))
my_df2
```
