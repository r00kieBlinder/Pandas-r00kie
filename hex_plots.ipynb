```python
import pandas as pd 
import numpy as np
from numpy.random import randn
%matplotlib inline 

# Create DataFrame
my_df = pd.DataFrame(randn(500, 4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
# Hexplots
my_df.plot(kind="hexbin", x='Mon', y='Tues')
```

```python
# Hexplot gridsize
my_df.plot(kind="hexbin", x='Mon', y='Tues', gridsize=10)
```

```python
# Hexplot color
my_df.plot(kind="hexbin", x='Mon', y='Tues', gridsize=10, cmap="Greens")
```

```python
# Other Method 
my_df.plot.hexbin(x='Mon', y='Tues', C='Wed', gridsize=10, cmap="Greens")
```

```python
# Title 
my_df.plot.hexbin(x='Mon', y='Tues', C='Wed', gridsize=10, cmap="Greens", title="My Awesome Hex Plots!")
```

```python
# Grid 
my_df.plot.hexbin(x='Mon', y='Tues', C='Wed', gridsize=10, cmap="Greens", title="My Awesome Hex Plots!", grid=True)
```
