```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create a quick DataFrame
my_df = pd.DataFrame(randn(100, 4), columns=["Mon", "Tues", "Wed", "Thur"])
my_df
```

```python
# Histograms
my_df["Wed"].hist()
```

```python
# Histograms with smaller bins, 50, 20 
my_df["Wed"].hist(bins=20)
```

```python
# Another way without the grid bars 
my_df["Wed"].plot(kind="hist", bins=20)
```

```python
my_df["Wed"].hist(bins=20, grid=True, legend=True)
```
