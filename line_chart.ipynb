```python
import pandas as pd 
import numpy as np
from numpy.random import randn
%matplotlib inline 

# Create DataFrame
my_df = pd.DataFrame(randn(25, 4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
# Line Chart
my_df.plot(kind="line")
```

```python
# Line Chart Absolute
my_df.abs().plot(kind="line")
```

```python
# Line Width
my_df.abs().plot(kind="line", lw=5)
```

```python
# Line Single
my_df["Mon"].abs().plot(kind="line", lw = 3)
```

```python
# Line more than one 
my_df[["Mon", "Wed"]].abs().plot(kind="line")
```

```python
# Size 
my_df[["Mon", "Wed"]].abs().plot(kind="line", figsize=(12, 5))
```

```python
# Title
my_df[["Mon", "Wed"]].abs().plot(kind="line", figsize=(12, 5), title="My Awesome Line Chart")
```

```python
# Shadding 
my_df[["Mon", "Wed"]].abs().plot(kind="line", figsize=(12, 5), title="My Awesome Line Chart", alpha=0.3)
```

```python
# Legend Removal
my_df[["Mon", "Wed"]].abs().plot(kind="line", figsize=(12, 5), title="My Awesome Line Chart", legend=False)
```

```python
# Second Way 
my_df.plot.line()
```
