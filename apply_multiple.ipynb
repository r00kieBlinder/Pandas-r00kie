```python
import pandas as pd
import numpy as np
# Create Data 
stuff = {
    'Corporation':['Apple', 'Google', 'Meta', 'Apple', 'Google', 'Meta'],
    'Employees':['Fahim', 'April', 'Wes', 'Beth', 'Justin', 'Steph'],
    'Salary':[200, 220, 190, 130, 120, 150]
}
# Create DataFrame
my_df = pd.DataFrame(stuff)
my_df
```

```python
# Create Function
def times1000(x):
    return format(x*1000, ",d")
```

```python
# Use apply function
var = my_df["Salary"].apply(times1000)
```

```python
# Append to current DF
my_df["Salary"] = var
my_df
```

```python
# Function on multiple columns
def namer(x):
    return "r00kie: " + x
```

```python
# Apply to Two columns
my_df[["Corporation", "Employees"]].apply(namer)
```

```python
# Append to current dataframe
my_df[["Corporation", "Employees"]] = my_df[["Corporation", "Employees"]].apply(namer)
my_df
```
